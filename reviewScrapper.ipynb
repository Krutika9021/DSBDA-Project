```python
import requests
import bs4
```

```python
req1 = requests.get('https://api.allorigins.win/raw?url=https://www.amazon.in/Apple-Headphones-Cancellation-Transparency-Personalised/dp/B0DGJ6G1XG/ref=sr_1_1')
req1
```

```python
req1.content
```

```python
soup = bs4.BeautifulSoup(req1.text)
```

```python
soup
```

```python
reviews = soup.find_all('div',{'class' : 'a-row a-spacing-small review-data'});
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
ratings = soup.find('div',{'class':'a-fixed-left-grid-col aok-align-center a-col-right'}).get_text()
print(ratings)
```

```python
individual_ratings = soup.find('div',{'class':'a-letter-space'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text() + "\n")
```

```python

individual_ratings = soup.find('div',{'class':'a-row'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text() + "\n")
```

```python
customer_name = soup.find_all('span',{'class':'a-profile-name'});
for cust_name in customer_name:
    print(cust_name.get_text() + "\n")
```
